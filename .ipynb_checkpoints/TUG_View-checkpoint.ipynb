```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
df = pd.read_csv('data/TEST01.csv', sep=',', skiprows=1, header=None)
```

```python
acc_x = df[1]
acc_y = df[2]
acc_z = df[3]
time = df[0]
```

```python
fig = go.Figure()

config = dict({'scrollZoom': True})

fig.add_trace(go.Scatter(x=time, y=acc_x,
                    mode='lines',
                    name='X Acceleration'))
fig.add_trace(go.Scatter(x=time, y=acc_y,
                    mode='lines',
                    name='Y Acceleration'))
fig.add_trace(go.Scatter(x=time, y=acc_z,
                    mode='lines',
                    name='Z Acceleration'))

fig.show(config=config)
```
