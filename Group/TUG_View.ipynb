```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import requests
import json
```

```python
url = "http://3.142.121.246:3000/getData?limit=100&skip=0"
 
response = requests.get(url)
print(response)
data = json.loads(response.text)
data = pd.DataFrame(data)
print(data)
```

```python
acc_x = data["column1"]
acc_y = data["column2"]
acc_z = data["column3"]
time = data["date_created"]
dataid = data["id"]
print(time)
```

```python
fig = go.Figure()

config = dict({'scrollZoom': True})

fig.add_trace(go.Scatter(x=time, y=acc_x,
                    mode='lines',
                    name='X Acceleration'))
fig.add_trace(go.Scatter(x=time, y=acc_y,
                    mode='lines',
                    name='Y Acceleration'))
fig.add_trace(go.Scatter(x=time, y=acc_z,
                    mode='lines',
                    name='Z Acceleration'))

fig.show(config=config)
```
